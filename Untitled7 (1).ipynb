Task 1.3 Otsu Thresholding function: Implement a function that applies Otsu's algorithm to calculate optimal threshold for image
Problem solving strategy and way of thinking:
To implement the Otsu Thresholding algorithm, I started by following the skeleton program from Canvas to structure my program

Problem solving strategy: Load and convert the image to grayscale,then calculate the histogram of pixel intensity from (0-255) and in the end
use a Otsu's thresholding algorithm to find the optimal threshold.
I used the skeleton program, it provided me with a framework for function like 
otsu_1(image) this function implements otsu method based on framework, this algorithm will calculate the between-class variance
for each of the threshold from (0-255) and will then choose the value that will maximizes variance.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

# Start by loading image by using backslahsed in the filepath
image_path = "C:\\Users\\abdio\\Downloads\\gingerbreads.jpg"

# Start by step1: load the image
image = io.imread(image_path)

# Step 2: Convert to grayscale if the image is colored (3D array)
if len(image.shape) ==3:
    image = image.mean(axis=2)

# Step 3: Then we compute the histogram
def c_histogram(image):
    histogram = np.zeros(256) # Create an array with 256 bins for each intensity value
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            pixel_value = int(image[i, j])
            histogram[pixel_value] += 1
    return histogram

# Step 4: Apply Otsu thresholding function
def otsu1(image):
    """ We will start to find the optimal threshold value using Otsu's method"""
    hist1 = c_histogram(image) 
    total_pixels = image.shape[0] * image.shape[1]

    current_max_variance_1 = 0 # Set max variance
    optimal_threshold_1 = 0    # Set Optimal threshold
    sum_total_1 = np.sum([i * hist[i] for i in range (256)]) # Total pixel intensities by histogram values
    sum_background_1 = 0 # Establish background pixel intensity sum
    weight_background_1 = 0 # Set background pixel weight 

    # Try all threshold values (0-255)
    for threshold in range(256):
        weight_background_1 += hist1[threshold] # Adjust the background weight 
        if weight_background_1 == 0: # Without background pixels, skip
            continue 
        
        weight_foreground_1 = total_pixels - weight_background_1 # Weight of foreground
        if weight_foreground_1 == 0: # If there are any pixels in the center, just stop
            break

        sum_background_1 += threshold * hist[threshold]
        mean_background_1 = sum_background_1 / weight_background_1
        mean_foreground_1 = (sum_total_1 - sum_background_1) / weight_foreground_1

        # Between class variance

        between_class_variance1 = weight_background_1 * weight_foreground_1 * (mean_background_1 - mean_foreground_1) ** 2

        if between_class_variance1 > current_max_variance_1:
            current_max_variance_1 = between_class_variance1
            optimal_threshold = threshold
    
    return optimal_threshold # This will return the best threshold


# Step 5: Compute the histgram and Otsu threshold
hist = c_histogram(image)
threshold_value = otsu1(image)

# Display the original grayscale image
plt.imshow(image, cmap='gray')
plt.title("Original Grayscale Image")
plt.show()

# Display the Histogram
plt.plot(hist)
plt.title("Histogram of the Image")
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of Pixels')
plt.show()

# Print the calculated Otsu threshold
print(f"The optimal threshold value calculated by Otsu's method is: {threshold_value}")
```
